```python
import json
import numpy as np
import pandas as pd

import csv
import ujson
```

```python
# json 파일 열어서 dataframe 형태로 변환
with open('D:/amazon sports/metadata.json') as f:
    js = json.loads(f.read())
df = pd.DataFrame(js)

# csv 파일로 저장
df.to_csv('D:/amazon sports/metadata_dataframe.csv')
```

```python
# csv 파일 읽기
df_csv = pd.read_csv('D:/amazon sports/metadata.csv')
df_csv
```

```python
# 행 별로 결측값 퍼센트 측정

for col in df_csv.columns:
    pct_missing = df_csv[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
# Nan 값이 대부분인 필요없는 행 삭제

df_csv = df_csv.drop(['tech1'], axis = 1) 
df_csv = df_csv.drop(['fit'], axis = 1) 
df_csv = df_csv.drop(['tech2'], axis = 1)
df_csv = df_csv.drop(['similar_item'], axis = 1)
# df_csv = df_csv.drop(['date'], axis = 1)

# price 값이 nan인 상품 삭제
df_csv = df_csv.dropna(subset=['price'], how='any', axis=0)
```
